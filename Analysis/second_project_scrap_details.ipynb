```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from concurrent.futures import ThreadPoolExecutor
import requests
import csv

def fetch_laptop_details(url):
    # Initialize WebDriver
    service = Service('C:/chromedriver-win64/chromedriver.exe')
    driver = webdriver.Chrome(service=service)
    
    # Open URL
    driver.get(url)
    
    # Wait for the page to load
    driver.implicitly_wait(10)

    # Initialize an empty dictionary to hold laptop details
    laptop_details = {}
    
    # Fetch Color using Selenium
    try:
        color_value = driver.find_element(By.CSS_SELECTOR, '#colorsWrapper .title-and-tick').text
    except:
        color_value = 'N/A'
    laptop_details['Color'] = color_value

    # Initialize variables to hold the prices
    original_price_value = None
    discount_price_value = None
    
    # Fetch Original Price using Selenium
    try:
        original_price_value = driver.find_element(By.CSS_SELECTOR, '.price-value-wrapper #lblOriginalPrice').text
    except:
        pass
    
    # Fetch Discount Price using Selenium
    try:
        discount_price_value = driver.find_element(By.CSS_SELECTOR, '.price-value-wrapper #lblPrice').text
    except:
        pass
    
    # If only one price is available, set both Original and Discount Price to that value
    if original_price_value and not discount_price_value:
        discount_price_value = original_price_value
    elif discount_price_value and not original_price_value:
        original_price_value = discount_price_value
    
    # If no prices are found, set them to 'N/A'
    if not original_price_value:
        original_price_value = 'N/A'
    if not discount_price_value:
        discount_price_value = 'N/A'
    
    laptop_details['Original Price'] = original_price_value
    laptop_details['Discount Price'] = discount_price_value
    laptop_details['URL'] = url  # Add URL to CSV
    
    # Use BeautifulSoup for other static details
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    specs_div = soup.select_one('.spec-wrapper')
    if specs_div:
        for feature_div in specs_div.select('.attribute-caption'):
            feature_name = feature_div.text.strip()
            feature_value_div = feature_div.find_next_sibling('div', class_='attribute-value')
            feature_value = feature_value_div.text.strip() if feature_value_div else 'N/A'
            laptop_details[feature_name] = feature_value
    
    # Close the browser
    driver.quit()
    
    print(f"Fetched details for {url}")
    return laptop_details

def main():
    # Read URLs from 'links.csv'
    with open('links.csv', 'r') as f:
        reader = csv.reader(f)
        urls = [row[0] for row in reader if not row[0].startswith('/myprofile')]
 
    # Add prefix to URLs
    urls = [f"https://meghdadit.com{url}" for url in urls]

    # Initialize ThreadPoolExecutor
    with ThreadPoolExecutor(max_workers=10) as executor:
        results = list(executor.map(fetch_laptop_details, urls))

    # Write to 'laptop_details.csv'
    with open('laptop_details.csv', 'w', newline='', encoding='utf-8') as f:
        writer = csv.DictWriter(f, fieldnames=sorted(set(key for laptop in results for key in laptop.keys())))
        writer.writeheader()
        writer.writerows(results)

if __name__ == "__main__":
    main()
```
