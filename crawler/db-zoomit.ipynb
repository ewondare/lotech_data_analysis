```python
session = Session(bind=engine)
```

```python
manufacturer = Table(
    'manufacturer' , meta,
    Column('id' , Integer , primary_key=True),
    Column('name' , String(200)),
    extend_existing=True,
) 
meta.create_all(engine)
session.commit()
```

```python
#it should run after other block
import pandas as pd
df = pd.read_csv ('Data/zoomit.csv' , encoding='UTF-8')

storage_query = "SELECT * FROM storage;"
product_query = "SELECT * FROM product;"
storage_df = pd.read_sql_query(storage_query, engine)
product_df = pd.read_sql_query(product_query, engine)

storage = df[['url', 'Storage', 'Storage_type']].drop_duplicates().reset_index(drop=True)
array = []
pattern = r'\((.*?)\)'
pattern2 = r'\b\d+\b'
storage["Storage"] = storage["Storage"].astype(str)

for i in range(len(storage)):
    url = storage["url"][i]
    if storage["Storage"][i] != None and storage["Storage"][i] != None:
        if "SSD" in storage["Storage"][i] and "HDD" in storage["Storage"][i]:
            if "(" in storage["Storage"][i] and ")" in storage["Storage"][i]:
                res_search = re.search(pattern, storage["Storage"][i])
                res_array = res_search.group(1).split("و")
                numbers_HDD = re.findall(pattern2, res_array[0])
                numbers_SSD = re.findall(pattern2, res_array[1])
                array.append([url, "HDD", numbers_HDD])
                array.append([url, "SSD", numbers_SSD])
            else:
                res_array = storage["Storage"][i].split("و")
                numbers_HDD = re.findall(pattern2, res_array[0])
                numbers_SSD = re.findall(pattern2, res_array[1])
                array.append([url, "HDD", numbers_HDD])
                array.append([url, "SSD", numbers_SSD])
        else:
            numbers = re.findall(pattern2, storage["Storage"][i])
            array.append([url, storage["Storage_type"][i], numbers])            
df = pd.DataFrame(columns=['url', 'type', 'size']) 
for ar in array:
    if ar[1] != None and len(ar[2]) > 0:
        df.loc[len(df)] = [ar[0], ar[1], int(ar[2][0])]
df = df.reset_index(drop=True)   
product_storage_df = pd.DataFrame(columns=["id_product", "id_storage"])
array = []
for i in range(len(df)):
    pro_id = int(product_df.loc[product_df['url'] == df['url'][i]]['id'])
    matching_rows = storage_df.loc[(storage_df['type'] == df['type'][i]) & (storage_df['size'] == df['size'][i]), 'id']

    if not matching_rows.empty:
        sto_id = int(matching_rows.iloc[0])
    else:
        sto_id = None 
    
    product_storage_df.loc[len(product_storage_df)] = [pro_id, sto_id]
product_storage_df.to_sql('product_storage' , con=engine ,  if_exists='append' , index=False)
```

```python
pattern2 = r'\b\d+\b'

manufacturer_query = "SELECT * FROM manufacturer;"
screen_size_query = "SELECT * FROM screen_size;"
resolution_query = "SELECT * FROM resolution;"
cpu_query = "SELECT * FROM cpu;"
ram_query = "SELECT * FROM ram;"
ram_type_query = "SELECT * FROM ram_type;"
gpu_query = "SELECT * FROM gpu;"
gpu_ram_query = "SELECT * FROM gpu_ram;"
storage_query = "SELECT * FROM storage;"
manufacturer_df = pd.read_sql_query(manufacturer_query, engine)
screen_size_df = pd.read_sql_query(screen_size_query, engine)
resolution_df = pd.read_sql_query(resolution_query, engine)
cpu_df = pd.read_sql_query(cpu_query, engine)
ram_df = pd.read_sql_query(ram_query, engine)
ram_type_df = pd.read_sql_query(ram_type_query, engine)
gpu_df = pd.read_sql_query(gpu_query, engine)
gpu_ram_df = pd.read_sql_query(gpu_ram_query, engine)
storage_df = pd.read_sql_query(storage_query, engine)

products_df = pd.DataFrame(columns=["url", "title", "model", "weight", "price", "score", "size", "battery_life", "manufacturer_id", "screen_size_id", "resolution_id", "cpu_id", "ram_id", "ram_type_id","gpu_id","gpu_ram_id", ])
product_storage_df = pd.DataFrame(columns=["id_products", "id_storage"])

df["Weight"] = df["Weight"].astype(str)
df["Price"]  = df["Price"].astype(str)
df["Screen_Size"]  = df["Screen_Size"].astype(str)
for i in range(len(df)):
    Weight = re.search(r'\d+\.\d+', df["Weight"][i])
    if Weight != None:
        new_Weight = float(Weight.group())
    else:
        new_Weight = None
    Price = df["Price"][i].replace(",", "").replace(" ", "")
    if Price.lower() == 'nan':
        Price = 0
    else:
        Price = int(float(Price))
    if df["Model_Name"][i] == None:
        Model_Name = None
    else:
        Model_Name = manufacturer_df.loc[manufacturer_df['name'] == df["Model_Name"][i]].values
    if (len(Model_Name) > 0):    
        new_Model_Name = int(Model_Name[0][0])
    else:
        new_Model_Name = None
        
    screen_size = re.search(r'\d+\.\d+', df["Screen_Size"][i])
    if screen_size != None:
        new_screen_size = screen_size.group()
        new_screen_size = screen_size_df.loc[screen_size_df['size'] == new_screen_size].values[0][0]
    else:
        new_screen_size = None
        
    if df["Resolution"][i] != None:
        Resolution = resolution_df.loc[resolution_df['name'] == df["Resolution"][i]].values
    else:
        Resolution = None
    
    if Resolution is not None and np.any(Resolution):
        new_Resolution = Resolution[0][0]
    else:
        new_Resolution = None
        
    if df["CPU"][i] != None:
        cpu = cpu_df.loc[cpu_df['name'] == df["CPU"][i]].values[0][0]
    else:
        cpu = None
    
    if df["RAM"][i] != None:
        ram = ram_df.loc[ram_df['name'] == df["RAM"][i]].values[0][0]
    else:
        ram = None
        
    if df["RAM_type"][i] != None:
        RAM_type = ram_type_df.loc[ram_type_df['type'] == df["RAM_type"][i]].values
    else:
        RAM_type = None
    
    if RAM_type is not None and np.any(RAM_type):
        new_RAM_type = RAM_type[0][0]
    else:
        new_RAM_type = None
        
    if df["GPU"][i] != None:
        gpu = gpu_df.loc[gpu_df['type'] == df["GPU"][i]].values
    else:
        gpu = None
    
    if gpu is not None and np.any(gpu):
        new_gpu = gpu[0][0]
    else:
        new_gpu = None 
        
    if df["GPU_RAM"][i] != None:
        gpu_ram = gpu_ram_df.loc[gpu_ram_df['size'] == df["GPU_RAM"][i]].values
    else:
        gpu_ram = None
    
    if gpu_ram is not None and np.any(gpu_ram):
        new_gpu_ram = gpu_ram[0][0]
    else:
        new_gpu_ram = None 

    products_df.loc[len(products_df)] = [
        df["url"][i],
        df["Title"][i],
        df["Manufacturer"][i],
        new_Weight,
        Price,
        float(df["Score"][i]),
        df["Size"][i],
        df["Battery_life"][i],
        new_Model_Name,
        new_screen_size,
        new_Resolution,
        cpu,
        ram,
        new_RAM_type,
        new_gpu,
        new_gpu_ram,
        ]
products_df.to_sql('product' , con=engine ,  if_exists='append' , index=False)

```
